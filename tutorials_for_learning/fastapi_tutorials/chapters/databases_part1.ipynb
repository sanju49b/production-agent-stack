```python
'''
A database model in SQLAlchemy is a python class that represents a table in the relational database, Instead of writing raw sql to create table
we define python classes that inherit from Base class provided by SQLAlchemy declarative
system, each attribute mapes to a clum in the table, this approach is called ORM

'''

from sqlalchemy import Column,Integer,String
from sqlalchemy.ext.declarative import declarative_base

#base is the foundation class all models must inherit from

Base = declarative_base()

# class user is defining from base which is declarative_base so all the properities of that clas
# will be applicable for the user clas here
'''
What is index=True mean

An index is a data structure the database builds with the bg to make lookups on a specific
column much faster, without an index, the database scans every row to find a match, with index=True, the database
builds a sorted index on that column so it can find rows in logarithmic time

Row Id Username Email
1   1  sai      sai@gmail.com

email index might look like

this is in
sai@gmail.com ---> row1

user index:
alex → row 3
john → row 2
sai → row 1

'''
class User(Base):
  __tablename__ = "users" #this is the actual table name in the db
  id = Column(Integer,primary_key=True,index=True)
  username = Column(String,unique=True,index=True)
  email = Column(String,unique=True,index=True)
```

```python
'''

create_all()

Once the model classes are defined , they only exist as python code . the actual tables dont exist in the database yet
you need to call Base.metadata.create_all(bind=engine) to physically create the tables in the database
SQLAlchemy inspects all classes that inherit from Base, reads their column definations and generates the corresponding table
create table sql statements, importantly this method is safe to call multiple times -- it checks whether the table exists before creating it,
so it will never overwrite or duplicate an existing table

'''

from sqlalchemy import create_engine
from sqlalchemy.ext.declarative import declarative_base
Base = declarative_base()
#base is not a table, it is not a database, it is simply and empty registry
# a container that will keep track of all your models
# this is like an empty notebook that SQLAlchemy will use to remember what tables need to exist

# Step 1: Define a model — this tells SQLAlchemy what the table should look like
class User(Base):
    __tablename__ = 'users'                          # table name in the database
    id = Column(Integer, primary_key=True, index=True)
    username = Column(String, unique=True, index=True)
    email = Column(String, unique=True, index=True)


# Define the database URL — SQLite stores the DB as a local file
SQLALCHEMY_DATABASE_URL = "sqlite:///./test.db"

engine = create_engine(
    SQLALCHEMY_DATABASE_URL,
    connect_args = {"check_same_thread":False}
)


'''
1. The registry containing all the model blue prints
2. Loop through every blueprint in the registry
3. use the engine to connect to the actual database
4. Check if the table already exists in the database
5. if not exisits ---> run the create table


'''
Base.metadata.create_all(bind=engine)


```
