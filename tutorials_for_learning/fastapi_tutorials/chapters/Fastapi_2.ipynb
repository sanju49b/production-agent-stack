#### Fundamentals of building apis

Defining routes

Fastapi makes it relatively simple to define api routes using decorators that coresspond to HTTP methods like GET,POST,PUT,DELETE, each method is used for different puprose when interacting with the server

```python
'''
GET method

the get method is used to retrieve information from the server it is the most
common method for fetching data, in fastapi you define get route usinf app.get()
decorator
'''
from fastapi import FastAPI
app = FastAPI()

@app.get("/items/{item_id}")
def read_item(item_id: int):
  return {"item_id": item_id}
```

```python
'''
post method

1. this is used to send data to the server, to create new resources, it is typically used for
submitting form data, creating new records in a database

2. we @app.post() decorator

'''
from pydantic import BaseModel
#defint item model
class Item(BaseModel):
  name:str
  price:int


@app.post("/items/")
def create_item(item:Item):
  return {"name":item.name,"price":item.price}
```

```python
'''
PUT method

this is to update the existing resource, it replaces the current representation of a resource
with the new one. in fastapi you define a put route using the @app.put()

'''

@app.put("/items/{item_id}")
def update_item(item_id:int,item:Item):
  return {"item_id":item_id,"name":item.name,"price":item.price}
```

```python

'''
The delete methos is used to delete the resources
'''

@app.delete("/items/{item_id}")
def delete_item(item_id: int):
  return {"message": f"Item with id {item_id} has been deleted"}
```

#### Path Parameters and Query Parameters

A path parameter is a part of the url path itself and is used to identify a specific resource - like saying give me item number 42 for example, in /items/42 , the 42 is the path parameter, and the fastapi passes it directly into your function as item_id, On the other hand query parameters are like optial settings or filters that come after a ? in url such as /items/?skip=5&limit=20, here you are not asking for one speicific item rather controlling how data is returned, may be skipping the first 5 items and returning or limiting result to 20, these are useful for features like pagination,searching or sorting, when you combine both, it even becomes powerful the path parameter identifies what you want eg item no 42 and query parameter identifies how much or what kind of details you want /item/45?detail=true to get the full details



```python
'''

Now that we have clear idea on path parameters and query parameters lets look at an example

#1. as discuessed path parameters are defined within {}, these are typically used to identify specific resources
#2. The value of the item_id is passed directly into th url
#3. Query parameters are optional and passed after a ?
#4. In this example, the route /items/ accepts two queries a skip and limit query ig they are not provided
with values they automatically become 0 and 10 respectively

The request might look something like this

GET /items/?skip=5&limit=20

#5. combining path and query parameters
GET /items/42?detail=true this is how the request might look
'''

@app.get("/item/{item_id}") #1
def read_item(item_id:int):
  return {"item_id":item_id} #2

@app.get("/items/") #3
def read_items(skip:int=0,limit:int=10): #4
  return {"skip":skip,"limit":limit}

#5
@app.get("items/{item_id}")
def read_item(item_id:int,detail:bool=False):
  if detail:
    return {"item_id":item_id,"detail":"Full item details"}
  return {"item_id":item_id}



```
