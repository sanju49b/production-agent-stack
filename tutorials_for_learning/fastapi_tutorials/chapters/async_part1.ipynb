#### Improving Performance with async

1. One of the standout features of FastAPI is its built-in support for asynchronous programming.

2. Asynchronous programming allows you to write non blocking code that can handle multiple tasks concurrenly, making the applications more efficient and scalable



#### Difference between Synchronous vs Asynchronous Operations

In traditional synchronous programming, each task is executed one after the other, blocking the execution of subsequent tasks until the current one is finishing, which leads to waiting such as database queries, file I/O or external api calls

Aysnchronous programming, on the other hand, allows you to write code that does not block the execution of other tasks while waiting for I/O bound operations to complete

with async programming,tasks that involve waiting like making HTTP requests or quering a db are executed concurrently allowing other tasks to processed in meantime

in python asynchronous programming is achieved usinfg the asyncio library, and the async def and await keywords are used to degine and execute asynchronous functions

```python
'''
Example of synchronous function

1. fecth_data and process_data are synchronous functions
2. Each function block the programme until it complete its tasks

'''

import time

def fetch_data():
  #simulation a blocking I/O operation from a db
  time.sleep(2)
  return {"message":"Data fetched Successfully"}

def process_data():
  time.sleep(1) #processing data just a simulation
  return {"message":"Data processed Successfully"}

def main():
  result1 = fetch_data()
  result2 = process_data()
  print(result1)
  print(result2)

s = main()

```

```python
'''
we define the endpoint functions with async def and FastAPI can hanlde multiple requests
concurrently without blocking,

this leads to improved performance, especially in scenarios involving I/O bound tasks

The await keyword in python is used inside an async def function to pause the current task until another async task finishes without blocking the entire program

The fake_db_query() function is defined with async def and uses await asyncio.sleep(2) to simulate a 2 second delay
tat might occur when querying a db or making an api request

The async_example() route is also defined using async def and awaits the result
of the fake_db_query() function.

Even though fake_db_query() has a 2 second delay , Fast API can continue processing
other requests, while waiting for the result of the query this allows the application to scale more effectively
by handling multiple reqyests concurrently without blocking the code

'''

from fastapi import FastAPI
import asyncio

app = FastAPI()

async def fake_db_query():
  await asyncio.sleep(2) #simulating a db query which would take 2 seconds to fetch
  return {"message":"Database query completed!"}

@app.get('/async-example')
async def async_example():
  result = await fake_db_query #awating the async database query
  return result
```

```python
'''

practical example of an async api

suppose we are building an api that interacts with 2 external services
one for fetching user data and another for retrieving transcation history
these operation might take time, so using asynchronous programming can help
imporce the perfromance of the api

The fetch_user_data and fetch_transcation_history() functions generally simulate
asynchronous I/O operations by using await aysncio.sleep() to represent the time

he get_user_info() route is asynchronous and calls both fetch_user_data()
and fetch_transaction_history(), awaiting both tasks concurrently.

sync the same operation would take 3 seconds while here it would take 2 seconds

'''
from fastapi import FastAPI
import asyncio

app = FastAPI()

async def fetch_user_data(user_id:int):
  await asyncio.sleep(1) #simulating a user id fetch
  return {"user_id":user_id,"name":"Sanjeev Bandaru"}

async def fetch_transaction_history(user_id:int):
  await asyncio.sleep(2) #simulating to fetch_transaction_history
  return {"user_id":user_id,"transcations":["purshase1","purashe 2"]}


@app.get("/user/{user_id}")
async def get_user_info(user_id:int):
  user_data = await fetch_user_data(user_id)
  transaction_data = await fetch_transaction_history(user_id)
  return {**user_data,**transaction_data}
```
