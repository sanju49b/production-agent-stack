#### Data validation

Pydantic is powerful data validation and parsing library, to handle all the incoming data, pydantic allows you to define data models that help ensure the correctness of data before it is used with your application

```python
'''
One of the most important features of pydantic is its ability to validate data
fastapi automatically validates the data aganist the model we define when handling requests
if the data is not matching then the fastapi will automaticall throw an error
'''

from fastapi import FastAPI
from pydantic import BaseModel

app  = FastAPI()

class Item(BaseModel):
  name: str
  price: float
  description: str = None
  tax:float = None

@app.post("/items")
async def create_item(item:Item):
  return {"name":item.name,"price":item.price}
```

```python
'''
Custom validation

we can also add validation logic, using pydantic's @root_validator or @validator decorators
these are used to apply additional checks or modify data before its returned

let say you want to make sure that text field cannot be negitive

in this case if you sent something like

{
"name":"Laptop",
"price": 1000,
"tax":-50
}
FastAPI will return an error response similar to:
{
"detail": [
{
"loc": ["body", "tax"],
"msg": "Sorry cannot be negitive",
"type": "value_error"
}
]
}



'''

from pydantic import BaseModel,validator

class Item(BaseModel):
  name: str
  price: float
  description: str = None
  tax:float = None

  @validator('tax')
  def validate_tax(cls,v):
    if v is not None and v < 0:
      raise ValueError('Sorry cannot be negitive')
    return v

```

```python
'''

Using Field() for default values and constraints

using this we can provide addtional validation for fields such as minimum and maximum values, length constraints or regular expressions
'''

from pydantic import BaseModel,Field
class Item(BaseModel):
  name: str = Field(..., min_length=3) # Required, with a minimum length of 3
  price: float = Field(..., ge=0) # Required, with a value greater than or equal to 0
  description: str = None # Optional field
  tax: float = Field(0.0, ge=0) # Optional field, with a default of 0.0 and a

```
