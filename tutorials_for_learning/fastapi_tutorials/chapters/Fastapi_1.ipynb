FastAPI and Uvicorn works together as cheif and waiter, the waiter takes the order and delivers food while fastapi the chef prepares it,

example:

1. user sends a request
2. uvicorn recieves the request first its the server listin on the port 8000
3. uvicorn passes the request to fastapi
4. fastapi porcess it
5. fastapi sends the response back to uvicorn
6. uvicorn sends it back to the user

```python
'''
running the first api
breakdown of the entire code with what does what

0. create a new python file:named main.py
0.1 . open the main.py file in the editor and write the followinf code

1. We are starting by importing fastapi package
2. we create instance of a class which represents the class
3. we are defining a route using @app.get("/") decorator, this route listent to
http get requests at the root URL
4. The read_root() function is the handler for this route, when the root url is accesses
it returns a dictonary with message: {"message":"Hello,FastAPI here"}

uvicorn main:app --reload

1. main refers to the python file name ggere
2. app refers to the created fastapi references
3. --reload: This option enables auto-reloading of the server whenever you make
changes to the code. This is useful during development

'''

from fastapi import FastAPI

#create an instance of the FastAPI class

app = FastAPI()

#define the route
@app.get("/")
def read_root():
  return {"message":"Hello,FastAPI here"}
```
